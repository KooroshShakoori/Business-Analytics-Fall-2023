# Jewelry Combination
For this problem we will first formulate the linear programming model.

## Defining the decision variables
The decision variables in this problem are the number of necklaces and bracelets we will be producing:
- N -> Number of Necklaces
- B -> Number of Bracelets

## Defining the Objective Function
considering that we want to maximize the profit and the profit for necklaces and bracelets are 300 and 400 respectively, the function is defined as below:

Maximize -> **300 * N + 400 * B**

## Defining Constraints
In this problem there are constraints on gold and platinum supplies and bracelet demands which is defined as follows:

- **3\*N + 2\*B <= 18, applies the constraint on the limited gold supply**
- **2\*N + 4\*B <= 20, applies the constraint on the limited platinum supply**
- **B <= 4, applies the constraint on the limited demand for bracelet**
- **B, N >= 0, applies the non-negativity constraints**

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(3, 1,figsize=(5, 15) ,layout='constrained')
axs[0].plot([0, 9],[6, 0], label='Gold Limit, slope = -2/3', color='yellow')
axs[0].plot([0, 5], [10, 0], label='Platinum Limit, slope = -2')
axs[0].plot([4, 4], [0, 10], label='Bracelet Demand')
axs[0].plot([0, 6], [8, 0], color='Red', linestyle='dotted', label='Objective Function, slope = -4/3')
axs[0].scatter([3], [4], color='Red')
axs[0].annotate(
    'Optimal solution', xy=(3, 4), xytext=(5, 6),
    arrowprops=dict(facecolor='r', edgecolor='k', shrink=0.05)
)
axs[0].set_xticks(list(range(11)))
axs[0].set_yticks(list(range(11)))
axs[0].set_xlabel('Bracelet Production')
axs[0].set_ylabel('Necklace Production')
axs[0].set_title('Graphical Solution')
axs[0].legend(loc=1)
axs[0].grid()

axs[1].plot([0, 9],[6, 0], label='Gold Limit, slope = -2/3', color='yellow')
axs[1].plot([0, 5], [10, 0], label='Platinum Limit, slope = -2')
axs[1].plot([4, 4], [0, 10], label='Bracelet Demand')
axs[1].plot([0, 9], [6, 0], color='Red', linestyle='dotted', label='Objective Function, slope = -2/3')
axs[1].set_xlabel('Bracelet Production')
axs[1].set_ylabel('Necklace Production')
axs[1].legend(loc=1)
axs[1].grid()
axs[1].annotate(
    '          Optimal solution line \nwhen Necklaces are sold for 600', xy=(5, 2.9), xytext=(4.1, 6),
    arrowprops=dict(facecolor='r', edgecolor='k', shrink=0.05)
)
axs[1].set_xticks(list(range(11)))
axs[1].set_yticks(list(range(11)))

axs[2].plot([0, 9],[6, 0], label='Gold Limit, slope = -2/3', color='yellow')
axs[2].plot([0, 5], [10, 0], label='Platinum Limit, slope = -2')
axs[2].plot([4, 4], [0, 10], label='Bracelet Demand')
axs[2].plot([0, 5], [10, 0], color='Red', linestyle='dotted', label='Objective Function, slope = -2')
axs[2].set_xlabel('Bracelet Production')
axs[2].set_ylabel('Necklace Production')
axs[2].legend(loc=1)
axs[2].grid()
axs[2].annotate(
    '          Optimal solution line \nwhen Necklaces are sold for 200', xy=(2.5, 5), xytext=(4.1, 6),
    arrowprops=dict(facecolor='r', edgecolor='k', shrink=0.05)
)
axs[2].set_xticks(list(range(11)))
axs[2].set_yticks(list(range(11)));
```
